```python
# Import libraries

import re
import time
import numpy as np
import pandas as pd
from sqlalchemy import create_engine
from pandasql import sqldf
from selenium.webdriver.support.ui import WebDriverWait
from selenium.webdriver.common.by import By
from selenium.webdriver.support import expected_conditions as EC
from selenium import webdriver
from selenium.webdriver.common.keys import Keys
from selenium.common.exceptions import NoSuchElementException
```

```python
#Importar los datos iniciales
data=pd.read_csv('st_221130_DA3.csv')
data.shape
```

```python
#UN muestreo random del 2%
data2=data.sample(frac=0.01, replace=True, random_state=1)
data2.shape
```

```python
#Define lista de urls
list_urls=data['Link']
#print(list_urls)
```

```python
#Abre el browser
browser=webdriver.Chrome("chromedriver.exe")
```

```python
counter=0
job_title=[]
company=[]

for url in list_urls:
    counter+=1
    browser.get(url)
    job1=browser.find_elements_by_class_name("Heading__H1-sc-uhbcp4-0") #Funciona para nombre del puesto
    job2=browser.find_elements_by_class_name("kWjXGu") #Funciona para nombre de compania
    #job3=browser.find_elements_by_class_name("dOmQK") #Funciona para job description
    for element in job1:
        job_title.append(element.text.lower())
        print(counter,"...",element.text.lower())
    for element in job2:
        company.append(element.text.lower())
                
print(len(job_title),len(company))
df=pd.DataFrame({"jpname":job_title,"company":company,"url":list_urls})
df.shape
```

```python
df.to_csv("st_221130_DA4.csv")
```

```python
# Drop special char in jpname
df['jd1']=df['jpname'].map(lambda x: re.sub(r'\W+',' ',x))
df['jd1']=df['jd1'].map(lambda x: re.sub(r'[0-9]+', '', x))
df['jd1']=df['jd1'].map(lambda x: re.sub(r' w ', '', x))
df['jd1']=df['jd1'].map(lambda x: re.sub(r' m ', '', x))
df['jd1']=df['jd1'].map(lambda x: re.sub(r' d ', '', x))
```
